A local variable is defined inside a function.

```python
def my_function():
    local_var = 10  # local variable
    print(local_var)  # Can be used only within this function

my_function()
# print(local_var)  # This will cause an error since local_var is not accessible here.
```

A global variable is defined outside of any function or block, typically at the top level of the script.

For Example:


```python
global_var = 20  # global variable

def my_function():
    print(global_var)  # Can access the global variable

my_function()
print(global_var)  # Can also access the global variable outside the function
```

To modify a global variable inside a function, use the global keyword.

For Example:

```python
global_var = 20  # global variable

def modify_global():
    global global_var  # Use global keyword to modify
    global_var = 30  # Modify the global variable

modify_global()
print(global_var)  # Now global_var is 30
```

### **1. Arithmetic operator**

Arithmetic operators are used with numeric values to perform common mathematical operations:

1. "+" (Addition): Adds two operands.
2. "-" (Subtraction): Subtracts the second operand from the first.
3. "*" (Multiplication): Multiplies two operands.

5. "%" (Modulus): Returns the remainder of the division.
6. "**" (Exponentiation): Raises the first operand to the power of the second.

```python
x = 10
y = 3

# Addition
addition_result = x + y
print(f"Addition: {addition_result}")

# Subtraction
subtraction_result = x - y
print(f"Subtraction: {subtraction_result}")

# Multiplication
multiplication_result = x * y
print(f"Multiplication: {multiplication_result}")

# Division
division_result = x / y
print(f"Division: {division_result}")

# Modulus
modulus_result = x % y
print(f"Modulus: {modulus_result}")

# Exponentiation
exponentiation_result = x ** y
print(f"Exponentiation: {exponentiation_result}")

# Floor Division
floor_division_result = x // y
print(f"Floor Division: {floor_division_result}")
```

### **Assignment operator**

  Assignment operators are used to assign values to variables:

  1. = (Assignment)

This is the basic assignment operator. It assigns the value on the right to the variable on the left.
- Example: x = 5 assigns the value 5 to the variable x.
2. += (Add and Assign)

This operator adds the value on the right to the variable on the left and assigns the result back to the variable.
- Example: x += 3 is equivalent to x = x + 3. If x was initially 5, it will become 8 after this operation.
3. -= (Subtract and Assign)

This operator subtracts the value on the right from the variable on the left and assigns the result back to the variable.
- Example: x -= 3 is equivalent to x = x - 3. If x was initially 5, it will become 2 after this operation.
4. *= (Multiply and Assign)

This operator multiplies the variable on the left by the value on the right and assigns the result back to the variable.
- Example: x *= 3 is equivalent to x = x * 3. If x was initially 5, it will become 15 after this operation.
5. /= (Divide and Assign)

This operator divides the variable on the left by the value on the right and assigns the result back to the variable.
- Example: x /= 3 is equivalent to x = x / 3. If x was initially 5, it will become approximately 1.6667 after this operation.
6. %= (Modulo and Assign)

This operator calculates the remainder when the variable on the left is divided by the value on the right and assigns the remainder back to the variable.
- Example: x %= 3 is equivalent to x = x % 3. If x was initially 5, it will become 2 after this operation (5 divided by 3 has a remainder of 2).
7. //= (Floor Division and Assign)

This operator performs floor division (division rounded down to the nearest integer) on the variable and the value, assigning the result back to the variable.
- Example: x //= 3 is equivalent to x = x // 3. If x was initially 5, it will become 1 after this operation (5 divided by 3, rounded down, is 1).
8. `=` (Exponentiation and Assign)**

This operator raises the variable on the left to the power of the value on the right and assigns the result back to the variable.
- Example: x **= 3 is equivalent to x = x ** 3. If x was initially 5, it will become 125 after this operation (5 raised to the power of 3 is 125).
9. &= (Bitwise AND and Assign)

This operator performs a bitwise AND operation between the variable and the value, assigning the result back to the variable.
- Example: x &= 3 is equivalent to x = x & 3. This operator works on the binary representations of the numbers.
10. |= (Bitwise OR and Assign)

This operator performs a bitwise OR operation between the variable and the value, assigning the result back to the variable.
- Example: `x |= 3` is equivalent to `x = x | 3`. This operator works on the binary representations of the numbers.
11. ^= (Bitwise XOR and Assign)

This operator performs a bitwise XOR operation between the variable and the value, assigning the result back to the variable.
- Example: `x ^= 3` is equivalent to `x = x ^ 3`. This operator works on the binary representations of the numbers.
12. ">>=" (Bitwise Right Shift and Assign)

This operator performs a bitwise right shift on the variable by the number of bits specified on the right, assigning the result back to the variable.
- Example: `x >>= 3` is equivalent

```python
# Basic Assignment
x = 10
print(f"x = {x}")

# Add and Assign
x += 5
print(f"x += 5: {x}")

# Subtract and Assign
x -= 3
print(f"x -= 3: {x}")

# Multiply and Assign
x *= 2
print(f"x *= 2: {x}")

# Divide and Assign
x /= 4
print(f"x /= 4: {x}")

# Modulo and Assign
x %= 5
print(f"x %= 5: {x}")

# Floor Division and Assign
x = 10  # Reset x to an integer
x //= 3
print(f"x //= 3: {x}")

# Exponentiation and Assign
x **= 2
print(f"x **= 2: {x}")

# Bitwise AND and Assign
x &= 5
print(f"x &= 5: {x}")

# Bitwise OR and Assign
x |= 3
print(f"x |= 3: {x}")

# Bitwise XOR and Assign
x ^= 2
print(f"x ^= 2: {x}")

# Bitwise Right Shift and Assign
x >>= 1
print(f"x >>= 1: {x}")

# Bitwise Left Shift and Assign
x <<= 2
print(f"x <<= 2: {x}")

# Walrus Operator (Assignment Expression)
print(f"x := 5: {(x := 5)}")
print(f"x: {x}")
```

### **Logical Operator**

Logical operators are used to combine conditional statements and produce a Boolean result (True or False) based on the logical relationship between the conditions.

Here are the three main logical operators in Python:

1. and (Logical AND)

This operator returns True if both conditions are True. If either condition is False, it returns False.
- Example: x > 5 and y < 10 returns True only if both x is greater than 5 and y is less than 10.
2. or (Logical OR)

This operator returns True if at least one of the conditions is True. It returns False only if both conditions are False.
- Example: x > 5 or y < 10 returns True if either x is greater than 5 or y is less than 10 (or both).
3. not (Logical NOT)

This operator reverses the result of a condition. If a condition is True, not makes it False, and vice versa.
- Example: not x > 5 returns True if x is not greater than 5.

```python
x = 10
y = 5

print(f"x > 5 and y < 10: {x > 5 and y < 10}")
print(f"x > 15 or y < 3: {x > 15 or y < 3}")
print(f"not x == y: {not x == y}")
```

### **Identity Operator**

Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

1. is

This operator returns True if both variables point to the same object in memory, otherwise False.
- Example: x is y returns True if x and y refer to the same object.
2. is not

This operator returns True if both variables do not point to the same object in memory, otherwise False.
- Example: x is not y returns True if x and y refer to different objects.



```python
# "Using is"
x = [1, 2, 3]
y = x  # `y` refers to the same object as `x`
z = [1, 2, 3]  # `z` is a different object with the same value

print(x is y)  # True, because `x` and `y` refer to the same object
print(x is z)  # False, because `x` and `z` are different objects
```

```python
# Using "is not"
x = [1, 2, 3]
z = [1, 2, 3]

print(x is not z)  # True, because `x` and `z` do not refer to the same object
```

```python
# Equality vs Identity:

"""The "==" operator checks if two objects have the same value.
The "is" operator checks if two objects are the same instance (memory location)."""

x = [1, 2, 3]
z = [1, 2, 3]
print(x == z)
print(x is z)
```

```python
"""
Immutable Types:
For immutable types like integers, strings, and tuples, Python may optimize memory
usage by referring to the same object for identical values"""
a = 10
b = 10
print(a is b)  # True, because small integers are cached

"""
Mutable Types:
For mutable types like lists and dictionaries, identical values are usually stored in separate objects.
"""
a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)  # False, because they are stored as separate objects

```

### **Membership Operator**

Membership operators are used to test if a sequence is presented in an object.

There are two membership operators.

1. in

This operator returns True if the value or variable is found in the sequence, otherwise False.
- Example: 'a' in 'banana' returns True because the letter 'a' is present in the string 'banana'.
2. not in

### **Bitwise Operator**


Bitwise operators work on the individual bits of integers. They perform operations on the binary representations of numbers.

Here are the bitwise operators in Python.
1. Bitwise And(&):

Sets each bit to 1 if both corresponding bits in the operands are 1.	x & y.	1


```python
x = 5  # 0b0101
y = 3  # 0b0011
result = x & y  # 0b0001
print(result)
```

2. Bitwise OR(|):

 Sets each bit to 1 if at least one corresponding bit in the operands is 1.	x | y.	7



```python
x = 5  # 0b0101
y = 3  # 0b0011
result = x | y  # 0b0111
print(result)
```

3. Bitwise XOR(^):

Sets each bit to 1 if only one corresponding bit in the operands is 1.	x ^ y.	6

```python
x = 5  # 0b0101
y = 3  # 0b0011
result = x ^ y  # 0b0110
print(result)
```

4. Bitwise NOT(~):

Inverts all the bits of the operand.

```python
x = 5  # 0b0101
result = ~x  # Inverts all bits: -(x+1)
print(result)
```

5. Zero fill left shift(<<)

Shifts the bits of the operand to the left by the specified number of positions.	x << 2.	20

```python
x = 5  # 0b0101
result = x << 2  # Shift left by 2 bits: 0b010100
print(result)
```

6. Signed right shift(>>):

Shifts the bits of the operand to the right by the specified number of positions.	x >> 2.	1

```python
x = 5  # 0b0101
result = x >> 2  # Shift right by 2 bits: 0b0001
print(result)
```

Join Sets:

There are several ways to join two or more sets in Python.

- The union() and update() methods joins all items from both sets.

- The intersection() method keeps ONLY the duplicates.

- The difference() method keeps the items from the first set that are not in the other set(s).

- The symmetric_difference() method keeps all items EXCEPT the duplicates.

1. Union():

The union() method returns a new set with all items from both sets.

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
union_set = set_1.union(set_2)
print(union_set)
```

- You can use the "|" operator instead of the union() method, and you will get the same result.

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
union_set = set_1 | set_2
print(union_set)

# Join multiple sets with the union() method
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = {5, 6, 7}
union_set = set_1.union(set_2, set_3)
print(union_set)


# Example 3
# Use | to join two sets
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = {5, 6, 7}
union_set = set_1 | set_2 | set_3
print(union_set)

# join a set with tuple
set_1 = {1, 2, 3}
tuple_1 = (4, 5, 6)
union_set = set_1.union(tuple_1)
print(union_set)
```

2. Update()

The update() changes the original set, and does not return a new set.

```python
set_1 = {"a","b","c"}
set_2 = {3, 4, 5}
set_1.update(set_2)
print(set_1)
```

The intersection() method will return a new set, that only contains the items that are present in both sets.

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)


"""You can use the & operator instead of the intersection() method, and you will get the same result."""
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 & set2
print(set3)

# intersection_update
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)
print(x)
```

4. Difference

The difference() method will return a new set that will contain only the items from the first set that are not present in the other set.

```python
# Keep all items from set1 that are not in set2
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)

print(set3)

# Example 2
"""You can use the - operator instead of
 the difference() method, and you will get the same result."""

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 - set2

print(set3)
```

```python
# Use the difference_update() method to keep the items that are not present in both sets
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.difference_update(set2)

print(set1)
```

5. Symmetric Differences

The symmetric_difference() method will keep only the elements that are NOT present in both sets.

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)
```

```python
# You can use the ^ operator instead of the symmetric_difference() method, and you will get the same result.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)
```

```python
# Use the symmetric_difference_update() method to keep the items that are not present in both sets
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.symmetric_difference_update(set2)

print(set1)
```

# **Python Lambda**
In Python, a lambda function is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but has only a single expression. The expression is evaluated and returned as the function's result.

Syntax:

- lambda arguments: expression

1. arguments: The inputs (parameters) to the function.
2. expression: The operation or computation that the function performs, and its result is returned.



```python
# Example
# Add 10 to argument a, and return the result
x = lambda a: a + 10
print(x(2))
```

```python
# Example 2
square = lambda x: x ** 2
print(square(5))
```

```python
# Example 3
# lambda with mupltiple arguments

multiply = lambda x, y: x * y
print(multiply(2, 3))
```

```python
# Example 4
# use that functions
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(3)

print(mydoubler(11))
```

### Key Features
1. Anonymous: Lambda functions are not bound to a name unless assigned explicitly.
2. Single Line: They consist of a single expression and are concise.
3. Temporary Use: Typically used for small tasks or as an argument to higher-order functions like map(), filter(), or sorted().

```python
# map(): Applies a function to each item in an iterable.

numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

```python
# filter(): Filters items in an iterable based on a condition.
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

```python
# sorted(): Custom sorting using lambda.
students = [("Alice", 25), ("Bob", 22), ("Charlie", 20)]
# Sort by age
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)

# Another example
# Sort by name
sorted_students = sorted(students, key=lambda x: x[0])
print(sorted_students)

print(type(sorted_students))
```

**Why Use Lambda Functions?**

- The power of lambda is better shown when you use them as an anonymous function inside another function.
Say you have a function definition that takes one argument, and that argument will be multiplied with an unknown number.
- For short-lived functions that don’t need a name.
- When using higher-order functions (map, filter, etc.).
- For readability in simple cases.

# **Python Try Except**

In Python, the **try** and **except** blocks are used for handling exceptions (runtime errors) gracefully. They allow you to catch errors in your code and define how to handle them, preventing your program from crashing.

- Syntax:

try:
    # Code that may raise an exception
except ExceptionType:
    # Code to handle the exception

- How It works

1. **try Block:** Contains the code that might raise an exception.
2. **except Block:** Handles the exception if it occurs. You can specify the type of exception to catch (e.g., ValueError, ZeroDivisionError).
3. **Optional else Block:** Executes if no exceptions occur in the try block.
4. **Optional finally Block:** Always executes, whether an exception occurs or not. Useful for cleanup code.

### Exception Handling

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.
These exceptions can be handled using the "try" statement:

```python
# The try block will generate an exception, because x is not defined
try:
  print(x)
except:
  print("An exception occurred")
```

Since the try block raises an error, the except block will be executed.
### Many Exceptions
You can define as many exception blocks as you want, e.g. if you want to execute a special block of code for a special kind of error:

```python
# Print one message if the try block raises a NameError and another for other errors
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

```python
# Example 3
# Print one message if the try block raises a NameError and another for other errors
try:
  print("Hello")
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

```python
# Example 4
try:
    x = 8
    result = 10 / x  # Might raise ZeroDivisionError
    print(f"Result: {result}")
except ValueError:
    print("Invalid input!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

### Else

You can use the else keyword to define a block of code to be executed if no errors were raised.

```python
# In this example, the try block does not generate any error
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")  # Always executes
```

### Finally
The finally block, if specified, will be executed regardless if the try block raises an error or not.

```python
x = 7
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")  # Always executes
```

```python
# Example 3
try:
    x = 6
    print(f"Square: {x ** 2}")
except ValueError:
    print("Please enter a valid number.")
else:
    print("Calculation successful!")  # Runs if no exception occurs
finally:
    print("Program ended.")  # Always executes
```

### Raise an exception
- As a Python developer you can choose to throw an exception if a condition occurs.

- To throw (or raise) an exception, use the raise keyword.

```python
# Example
# Raise an error and stop the program if x is lower than 0
x = -1

if x < 0:
  raise Exception("Sorry, no numbers below zero")
```

- The raise keyword is used to raise an exception.

- You can define what kind of error to raise, and the text to print to the user.

```python
# Example
# Raise a TypeError if x is not an integer
x = "hello"

if not type(x) is int:
  raise TypeError("Only integers are allowed")
```

#**Python String Formatting**

String formatting in Python allows you to create strings by embedding variables and expressions into predefined text. This makes dynamic string creation efficient and readable. Python offers multiple ways to format strings:

**1. Using the format() Method**

The format() method allows placeholders {} in a string to be replaced with values.

Syntax:


```python
"string with placeholders {}".format(values)
```

```python
# Example
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
```

```python
# Positional Arguments
print("{0} is learning {1}.".format("Alice", "Python"))
```

```python
# Keyword Arguments
print("{name} is {age} years old.".format(name="Bob", age=30))
```

```python
# Formatting Numbers
pi = 3.14159
print("Pi is approximately {:.2f}".format(pi))  # 2 decimal places
```

**2. Using f-Strings (Python 3.6 and Later)**

f-Strings are the most modern and concise way to format strings. They start with an f and allow variables or expressions to be directly embedded in curly braces {}.

Syntax:

```python
f"string with {variable_or_expression}"
```

```python
# Example 1
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
```

```python
# Example 2
# Expressions in f-Strings
a = 5
b = 10
print(f"The sum of {a} and {b} is {a + b}.")
```

**3. Using String Concatenation (Not Recommended)**

String concatenation with + can be used for formatting but is less efficient and harder to read for complex strings.

Example:

```python
name = "Alice"
age = 25
print("My name is " + name + " and I am " + str(age) + " years old.")
```

**4. Using Percent (%) Formatting**

This is an older method, similar to C-style string formatting.

Syntax:

```python
"string with %s or %d" % (values)
```

```python
# Example
name = "Bob"
age = 30
print("My name is %s and I am %d years old." % (name, age))
```

```python
# Example 2
# Formating number
pi = 3.14159
print("Pi is approximately %.2f" % pi)  # 2 decimal places
```

## Formatting Types

Alignment and Sign Formatting

```python
"""1)  :<
Left-aligns the result within the available space."""

print("{:<10}".format("Hello"))
```

```python
"""2) :>
Right-aligns the result within the available space."""

print("{:>10}".format("Hello"))
```

```python
"""3) :^
Centers the result within the available space."""

print("{:^10}".format("Hello"))
```

```python
"""4)  :=
Places the sign (+ or -) to the leftmost position in numbers."""

print("{:=+10}".format(123))
print("{:=+10}".format(-123))
```

**Thousand Separators**

```python
"""1)  , (Comma)
Use a comma as a thousand separator."""

print("{:,}".format(1000000))
```

```python
"""2)  :_
Use an underscore as a thousand separator."""

print("{:_}".format(1000000))
```

**Number Formats**

```python
"""1)  :b
Binary format (base 2)."""

print("{:b}".format(10))
```

```python
"""2)  :c
Converts the value into the corresponding Unicode character."""
print("{:c}".format(65))
```
